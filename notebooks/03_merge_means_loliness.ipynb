```python
import sys
from pathlib import Path
PROJECT_ROOT = Path.cwd().parent
sys.path.append(str(PROJECT_ROOT))
print("✅ Working directory set to project root:", PROJECT_ROOT)
```

```python
import pandas as pd
```

```python
DATAFOLDER = PROJECT_ROOT / "data"
EXTRACTED_FOLDER = DATAFOLDER / "extracted"
FINAL_FOLDER = DATAFOLDER / "final"
JSON_FOLDER = PROJECT_ROOT / "hackathon-app" / "public" / "data"
EXTRACTED_FOLDER.mkdir(parents=True, exist_ok=True)
FINAL_FOLDER.mkdir(parents=True, exist_ok=True)
JSON_FOLDER.mkdir(parents=True, exist_ok=True)
print("✅ Data folders set up at:", DATAFOLDER)
```

```python
means = pd.read_csv(EXTRACTED_FOLDER / "means_quarterly.csv")
lonely = pd.read_csv(EXTRACTED_FOLDER / "loneliness_quarterly.csv")

combined = means.merge(lonely, on=["quarter","geography_level", "geography"], how="left")
combined["year"] = 2025
combined["quarter"] = combined["quarter"].str.extract(r"Q(\d)").astype(int)
combined = combined[
    ["year","quarter","geography_level", "geography",
     "life_satisfaction","worthwhile","happiness","anxiety",
     "wellbeing_index",
     "often","some_time","occasionally","hardly_ever","never",
     "lonely_at_least_occasionally_pct",
     "lonely_often_or_sometimes_pct"]
]
```

```python
combined = combined.sort_values(
    ["year", "quarter", "geography_level", "geography"]
).reset_index(drop=True)


combined = combined.round(2)
```

```python
combined.to_json(
    JSON_FOLDER / "wellbeing_master.json",
    orient="records",
    indent=2
)
```

```python

combined.to_csv(FINAL_FOLDER / "means_loneliness_quarterly.csv", index=False)
print("✅ Combined data saved to:", FINAL_FOLDER / "means_loneliness_quarterly.csv")
```
